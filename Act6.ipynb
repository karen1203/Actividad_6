Remplazamos los valores nulos por las Ciudad de Mexico ya que es la más concurrida.

```python
mexico1['host_location']=mexico1['host_location'].fillna('Mexico City, Mexico')
```

Revisamos la moda de el tiempo de respuesta por ser una variable categórica.

```python
t2=freq_tbl(mexico1['host_response_time'])
t2
```

Lo reemplazamos por within an hour ya que por mucho es la más concurrida

```python
mexico1['host_response_time']=mexico1['host_response_time'].fillna('within an hour')
```

```python
t3=freq_tbl(mexico1['host_response_rate'])
t3
```

```python
mexico1['host_response_rato']=mexico1['host_response_rate'].fillna('100%')
```

```python
t4=freq_tbl(mexico1['neighbourhood'])
t4
```

```python
t5=freq_tbl(mexico1['host_is_superhost'])
t5
```

```python
mexico1['host_is_superhost']=mexico1['host_is_superhost'].fillna('f')
```

```python
t6=freq_tbl(mexico1['host_acceptance_rate'])
t6
```

```python
mexico1['host_acceptance_rate']=mexico1['host_acceptance_rate'].fillna('100%')
```

```python
mexico1["review_scores_cleanliness"]=mexico1["review_scores_cleanliness"].fillna(round(mexico1["review_scores_cleanliness"].mean(),1))
```

```python
mexico1["review_scores_value"]=mexico1["review_scores_value"].fillna(round(mexico1["review_scores_value"].mode(),1))
```

```python
mexico1['neighbourhood']=mexico1['neighbourhood'].fillna('Mexico City, Ciudad de México, Mexico')
```

En la siguiente variable, llenamos los valores nulos con el promedio.

```python
mexico1["review_scores_accuracy"]=mexico1["review_scores_accuracy"].fillna(round(mexico["review_scores_accuracy"].mean(),1))
```

```python
mexico.isnull().sum()
```

```python
cuali=mexico1[['host_name','host_since','host_location','host_response_time','host_response_rate','host_listings_count','host_verifications','host_identity_verified','neighbourhood','property_type','room_type','accommodates','price','minimum_nights','maximum_nights','number_of_reviews','host_is_superhost']]
```

```python
cuanti=mexico1[['review_scores_accuracy','host_acceptance_rate','review_scores_cleanliness','review_scores_value','availability_365']]
```

```python
plt.boxplot(cuanti['review_scores_accuracy'])
```

```python
q1 = cuanti['review_scores_accuracy'].quantile(0.25)
q2 = cuanti['review_scores_accuracy'].quantile(0.75)
iqr = q2 - q1

lower_bound1 = q1 - 1.5 * iqr
upper_bound1 = q2 + 1.5 * iqr
cuanti = cuanti[(cuanti['review_scores_accuracy'] >= lower_bound1) & (cuanti['review_scores_accuracy'] <= upper_bound1)]
```

```python
q11 = cuanti['review_scores_cleanliness'].quantile(0.25)
q21 = cuanti['review_scores_cleanliness'].quantile(0.75)
iqr1 = q21 - q11

lower_bound11 = q11 - 1.5 * iqr1
upper_bound11 = q21 + 1.5 * iqr1
cuanti = cuanti[(cuanti['review_scores_cleanliness'] >= lower_bound11) & (cuanti['review_scores_cleanliness'] <= upper_bound11)]
```

```python
q12 = cuanti['review_scores_value'].quantile(0.25)
q22 = cuanti['review_scores_value'].quantile(0.75)
iqr2 = q22 - q12

lower_bound12 = q12 - 1.5 * iqr2
upper_bound12 = q22 + 1.5 * iqr2
cuanti = cuanti[(cuanti['review_scores_value'] >= lower_bound12) & (cuanti['review_scores_value'] <= upper_bound12)]
```

```python
q13 = cuanti['availability_365'].quantile(0.25)
q23 = cuanti['availability_365'].quantile(0.75)
iqr3 = q23 - q13

lower_bound13 = q13 - 1.5 * iqr3
upper_bound13 = q22 + 1.5 * iqr3
cuanti = cuanti[(cuanti['availability_365'] >= lower_bound13) & (cuanti['availability_365'] <= upper_bound13)]
```

```python
mexlimp=pd.concat([cuanti,cuali], axis=1)
```
